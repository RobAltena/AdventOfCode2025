```python
import requests

data = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day2_input.txt').text

ranges = data.split(',')
```

```python
result = 0

for r in ranges:
  min_val, max_val = r.split('-')

  min_val_int = int(min_val)
  max_val_int = int(max_val)

  # let's slug our way to the minimum number we need to ckeck:
  if (len(min_val)//2 > 1):
    x1 = int(min_val[:len(min_val)//2]) -1
  else:
    x1 = 0

  while True:
    x1 = x1 + 1
    x = int(str(x1) + str(x1))
    if min_val_int <= x <= max_val_int:
      #print(f"   {x}")
      result += x
    else:
      if x > max_val_int:
        break

  #print(min_val, max_val, len(min_val)//2, len(max_val)//2, x)
  #print()

print(f'Day 2 Part 1: {result}')
```

```python

# Find the max number of repetitions we need to look for.
max_result = 0
for r in ranges:
  min_val, max_val = r.split('-')
  min_val_int = int(min_val)
  max_val_int = int(max_val)
  if max_val_int > max_result: max_result = max_val_int

#print(max_result)
max_reps = len(str(max_result))
#print(max_reps)
result = set()
for num_reps in range(2, max_reps+1):
  #print(num_reps)
  for r in ranges:
    min_val, max_val = r.split('-')

    min_val_int = int(min_val)
    max_val_int = int(max_val)

    # let's slug our way to the minimum number we need to ckeck:
    if (len(min_val)//num_reps > 1):
      x1 = int(min_val[:len(min_val)//num_reps]) -1
    else:
      x1 = 0

    while True:
      x1 = x1 + 1
      x = int(str(x1) * num_reps)
      if min_val_int <= x <= max_val_int:
        #print(f"   {x}")
        result.add(x)
      else:
        if x > max_val_int:
          break

print(f'Day 2 Part 2: {sum(result)}')
```
