```python
import numpy as np
import requests

data = """123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day6_input.txt').text


lines = data.split('\n')[:-1]

numberlines = []
for line in lines[:-1]:
    numberline = list(map(int, line.split()))
    numberlines.append(numberline)

operators = lines[-1].split()
```

```python
# To prevent drama: Check the lengths of the data.
l = len(numberlines[0])
for numberline in numberlines:
    if len(numberline) != l:
        print('Line length is not equal')

if len(operators) != l:
    print('Number of operators is not equal to number of lines')
```

```python
grand_total = 0
for col_index, op_symbol in enumerate(operators):
  total = 0 if op_symbol == '+' else 1
  for numberline in numberlines:
    if op_symbol == '+':
      total += numberline[col_index]
    else :
      total *= numberline[col_index]

  grand_total += total

print(f'Day 6, part 1: {grand_total}')
```

```python
# Transform lines into a character numpy array.
arr = np.array([list(line) for line in data.split('\n')[:-1]])

index = arr.shape[1] -1

pars = []
result = 0
while index >= 0:

    try:
      pars.append(int(''.join(arr[:-1, index])))
    except:
      pass

    if arr[-1, index] != ' ' :
      if arr[-1, index] == '+':
        result += sum(pars)
        #print(arr[-1, index], pars, sum(pars))
      else :
        result += np.prod(pars)
        #print(arr[-1, index], pars, np.prod(pars))
      pars = []

    index -= 1

print(f'Day 6, part 2: {result}')
```
