```python
import requests

data = """987654321111111
811111111111119
234234234234278
818181911112111"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day3_input.txt').text

banks = data.split()
```

```python
joltage = 0
for bank in banks:
  batteries = list(bank)

  max_bat = max(batteries[:-1])
  max_index = batteries.index(max_bat)

  max_bat2 = max(batteries[max_index+1:])
  joltage += int(max_bat + max_bat2)

print(f'Day 3, part 1: {joltage}')
```

```python
total_output_joltage = 0

for bank in banks:
 batteries = list(bank)

 l2 = batteries
 joltage = []
 for d in range(12, 0, -1):
  l3 = l2[:-(d-1)] if d > 1 else l2

  max_bat = max(l3)
  max_index = l3.index(max_bat)
  l2 = l2[max_index+1:]

  joltage.append(max_bat)

 #print(int("".join(joltage)))
 total_output_joltage += int("".join(joltage))

print(f'Day 3, part 2: {total_output_joltage}')

```
