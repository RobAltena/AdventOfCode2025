```python
import requests

data = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day1_input.txt').text
```

```python
# Part 1

pos = 50
counter = 0
for i in data.split():
  delta = int(i[1:])
  pos += delta if i.startswith('R') else -delta
  pos %= 100

  if pos == 0: counter += 1
  #print(i[0], i[1:], pos)

print(counter)
```

```python
def calcTicks(delta, pos):
    crossings = 0

    # Step 1: Calculate full rotations
    fullrot = abs(delta) // 100
    crossings += fullrot

    # Step 2: Adjust delta to be the remaining part after full rotations
    remaining_delta = delta
    if delta > 0:
        remaining_delta = delta - fullrot * 100
    elif delta < 0:
        remaining_delta = delta + fullrot * 100

    # Step 3: Check for boundary crossing due to the remaining delta
    if remaining_delta < 0 and pos != 0 and (pos + remaining_delta) <= 0:
        crossings += 1
    elif remaining_delta > 0 and (pos + remaining_delta) >= 100:
        crossings += 1

    # Step 4: Calculate the final position
    final_pos = (pos + delta) % 100
    # Python's % can return negative results for negative inputs, adjust if necessary
    if final_pos < 0:
        final_pos += 100

    return (final_pos, crossings)
```

```python
pos = 50
counter = 0
for i in data.split():
  delta = int(i[1:])
  delta = delta if i.startswith('R') else -delta

  (newpos, ticks) = calcTicks(delta, pos)
  pos = newpos
  counter += ticks

print(counter)
```
