```python
import requests

data = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day1_input.txt').text
deltas = list((int(i[1:]) if i.startswith('R') else -int(i[1:]) for i in data.split()))
```

```python
pos = 50
counter_part1 = 0
counter_part2 = 0
for delta in deltas:
  step = 1 if delta > 0 else -1
  for _ in range(abs(delta)):
    pos = (pos + step) % 100
    if pos == 0: counter_part2 +=1

  if pos == 0: counter_part1 +=1

print(f'Day 1, part 1: {counter_part1}')
print(f'Day 1, part 2: {counter_part2}')
```
