```python
import functools
import requests
import sys


data = """aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day11_input.txt').text

graph = {}

for line in data.split("\n")[:-1]:
    s, d = line.split(":")
    d = d.split()
    graph[s] = d

stack = ['you']
result = 0

while stack:
  node = stack.pop()
  d = graph[node]
  for n in d:
    if n == 'out':
      result += 1
    else:
      stack.append(n)


print(f'Day 11, part 1: {result}')
```

```python
graph['out'] = []

@functools.lru_cache(None)
def count_paths(node, visited_dac, visited_fft):
    if node == 'out':
        return 1 if visited_dac and visited_fft else 0

    total_paths = 0

    for neighbor in graph.get(node, []):
        next_dac = visited_dac or (neighbor == 'dac')
        next_fft = visited_fft or (neighbor == 'fft')

        total_paths += count_paths(neighbor, next_dac, next_fft)

    return total_paths


result = count_paths('svr', False, False)

print(f'Day 11, part 2: {result}')
```
