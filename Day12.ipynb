```python
import numpy as np
import requests

data = """0:
###
##.
##.

1:
###
##.
.##

2:
.##
###
##.

3:
##.
###
##.

4:
###
#..
###

5:
###
.#.
###

4x4: 0 0 0 0 2 0
12x5: 1 0 1 0 2 2
12x5: 1 0 1 0 3 2
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day12_input.txt').text


datablocks = data.split('\n\n')
shapes = datablocks[:-1]
regions = datablocks[-1].split('\n')[:-1]

SHAPE_DIR = {}

result = 0
for shape in shapes:
  lines = shape.split()
  id = int(lines[0][0])
  shape_arr = (np.array([list(line) for line in lines[1:]]) == '#').astype(int)
  SHAPE_DIR[id] = shape_arr

```

```python

result = 0
for r in regions:
  a, s = r.split(':')
  area = np.prod(list(map(int, a.split('x'))))
  shape_area = sum(map(int, s.split())) * 9  #This is an obvious hack. Why does it work?
  if area >= shape_area :
    result += 1

print(result)
```
