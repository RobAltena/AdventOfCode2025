```python
import requests

data = """3-5
10-14
16-20
12-18

1
5
8
11
17
32
"""

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day5_input.txt').text

ranges, ids = data.split('\n\n')

r = list([[int(num) for num in x.split('-')] for x in ranges.split()])

ids = list(map(int, ids.split()))

result = 0
for id in ids:

  for start, end in r:
    if start <= id <= end:
      result += 1
      break

print(f'Day 5, part 1: {result}')
```

```python
r.sort(key=lambda x: x[0])

merged_ranges = []

if not r:
  print("No ranges to merge.")
else:
  merged_ranges.append(list(r[0]))

  for current_start, current_end in r[1:]:
    last_merged_start, last_merged_end = merged_ranges[-1]

    if current_start <= last_merged_end:
      merged_ranges[-1][1] = max(last_merged_end, current_end)
    else:
      merged_ranges.append([current_start, current_end])

total_unique_ids = 0

for start, end in merged_ranges:
  total_unique_ids += (end - start + 1)

print(f'Day 5, part 2: {total_unique_ids}')
```
