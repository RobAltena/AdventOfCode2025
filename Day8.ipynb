```python
import numpy as np
from scipy.spatial.distance import cdist
import requests

np.set_printoptions(linewidth=200, threshold=np.inf)


data = """162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689
"""
i_max = 10

data = requests.get('https://raw.githubusercontent.com/RobAltena/AdventofCode2025/refs/heads/main/advent_day8_input.txt').text
i_max = 1000


arr = np.array([list(map(int, line.split(','))) for line in data.split()])
distance_matrix = cdist(arr, arr)
```

```python
distances_filtered = []
point_pairs_filtered = []

# Get the indices that would sort the flattened distance matrix
sorted_indices = np.argsort(distance_matrix.ravel())

# Iterate through the sorted indices to get distances and their corresponding point pairs
for flat_idx in sorted_indices:
    distance = distance_matrix.ravel()[flat_idx]
    row, col = np.unravel_index(flat_idx, distance_matrix.shape)

    # Only add if row < col to avoid duplicates (e.g., (0, 19) and (19, 0)) and self-distances (row == col)
    if row < col:
        distances_filtered.append(distance)
        point_pairs_filtered.append((row, col))

distances = distances_filtered
point_pairs = point_pairs_filtered

print("First 10 increasing distances (excluding zeros and duplicates):")
print(distances[:10])
print("\nFirst 10 corresponding point pairs (excluding zeros and duplicates):")
print(point_pairs[:10])
```

```python
circuits = []
for i in range(arr.shape[0]):
  circuits.append(set([i]))


connections = 0

for i in range(i_max):

  p = point_pairs[i]

  #Find the indices of the circuits that contain the points in p.
  circuit_indices = []

  for idx, circuit in enumerate(circuits):
      if p[0] in circuit or p[1] in circuit:
          circuit_indices.append(idx)

  if len(circuit_indices) == 2:
    circuits[circuit_indices[0]].update(circuits[circuit_indices[1]])
    circuits.pop(circuit_indices[1])
    connections += 1

l = list(map(lambda x: len(x), circuits))
l.sort()
print(f"Day 8, part 1: {np.prod(l[-3:])}" )
```
